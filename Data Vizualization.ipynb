```python
import os
# import librosa   #for audio processing
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
from pedalboard import Pedalboard
from pedalboard.io import AudioFile
from IPython.display import Audio
from collections import defaultdict
import time
```

### Visualize Audio signal

```python
train_audio_path = "data/"
audio_file = AudioFile("data/play_spotify1.mp3", 'r')
audio = audio_file.read(audio_file.frames)
samplerate = audio_file.samplerate
```

```python
fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1.set_title('Raw wave of ' + 'data/play_spotify1.mp3')
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
ax1.plot(np.linspace(0, samplerate/len(audio[0]), int(samplerate)), audio[0][:44100])
```

### Number of recordings for each command

```python
labels=os.listdir(train_audio_path)
```

```python
audio_types = [
    "increase_volume1.mp3",
    "increase_volume2.mp3",
    "next_song1.mp3",
    "next_song2.mp3",
    "pause_music1.mp3",
    "pause_music2.mp3",
    "play_spotify1.mp3",
    "play_spotify2.mp3",
    "reduce_volume1.mp3",
    "reduce_volume2.mp3",
    "skip_song1.mp3",
    "skip_song2.mp3",
    "stop_music1.mp3",
    "stop_music2.mp3",
]
```

```python
recordings = defaultdict(lambda: 0)
for label in labels:
    for t in audio_types:
        if (t in label):
            recordings[t] += 1
```

```python
plt.figure(figsize=(30,5))
index = np.arange(len(audio_types))
plt.bar(index, recordings.values())
plt.xlabel('Commands', fontsize=12)
plt.ylabel('No of recordings', fontsize=12)
plt.xticks(index, audio_types, fontsize=15, rotation=60)
plt.title('No. of recordings for each command')
plt.show()
```
