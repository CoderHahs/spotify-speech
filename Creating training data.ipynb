```python
from pedalboard import Pedalboard, Chorus, Reverb, PitchShift, Compressor
from pedalboard.io import AudioFile
from IPython.display import Audio
import time
```

```python
audio_file = AudioFile("data/play_spotify1.mp3", 'r')
audio = audio_file.read(audio_file.frames)
samplerate = audio_file.samplerate
    
board = Pedalboard([PitchShift(semitones=15)])

effected = board(audio, samplerate)

new_audio_file = AudioFile('data/processed-play_spotify1.wav', 'w', samplerate, effected.shape[0])
new_audio_file.write(effected)
```

```python
files = [
    "increase_volume1.mp3",
    "increase_volume2.mp3",
    "next_song1.mp3",
    "next_song2.mp3",
    "pause_music1.mp3",
    "pause_music2.mp3",
    "play_spotify1.mp3",
    "play_spotify2.mp3",
    "reduce_volume1.mp3",
    "reduce_volume2.mp3",
    "skip_song1.mp3",
    "skip_song2.mp3",
    "stop_music1.mp3",
    "stop_music2.mp3",
]
```

```python
for file in files:
    counter = 1
    start = time.time()
    print("starting", file, start)
    for c in [True, False]:
#         print(c)
        for r in range(5, 54, 5):
            r = r/100
            print(r)
            for p in range(-5, 5, 1):
#                 print(p)
                for d in range(-50, -4, 5):
                    
                    audio_file = AudioFile("data/"+file, 'r')
                    audio = audio_file.read(audio_file.frames)
                    samplerate = audio_file.samplerate
                    
                    if c:
                        board = Pedalboard([Chorus(), Reverb(room_size=r), PitchShift(semitones=p), Compressor(threshold_db=d)])
                    else:
                        board = Pedalboard([Reverb(room_size=r), PitchShift(semitones=p), Compressor(threshold_db=d)])

                    effected = board(audio, samplerate)
                    
                    new_file = "data/processed" + "_" + str(counter) + "_" + file[:-4] + ".wav"
                        
                    new_audio_file = AudioFile(new_file, 'w', samplerate, effected.shape[0])
                    new_audio_file.write(effected)
                    
                    counter += 1
    print("done", file, time.time()-start)
```
